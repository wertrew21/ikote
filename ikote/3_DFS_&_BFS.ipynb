# 큐
선입선출 구조

```
# 리스트 구조를 사용하여 큐를 구현할 수 있음.
# 그러나 삭제 시 시간복잡도가 O(1)보다 커지므로(O(k)) 덱을 이용하여 큐를 구현.

from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue) # output from first-come = <3 7 1 4<
queue.reverse()
print(queue) # output from last-come = >4 1 7 3>
```

# 재귀 함수 Recursive Function

```
# 무한히 반복되는 재귀함수. 내부 메모리가 가득 차면 오류가 뜨면서 종료.
def recursive_function():
    print('재귀 함수를 호출합니다.')
    recursive_function()

recursive_function()
```

```
# 그러므로 재귀함수의 종료 조건을 반드시 명시해야 함
def recursive_function(i):
    if i == 10:
        return
    print(i, '번째 재귀함수에서', i+1, '번째 재귀함수를 호출합니다.')
    recursive_function(i+1)
    print(i, '번째 재귀함수를 종료합니다.')

recursive_function(1)
```

위와 같이 재귀함수는 def recursive_function(i) ~ recursive_function(i+1) \
부분까지 반복해서 호출한 후(1-->9), 종료 조건(10)에 다다르면 \
recursive_function(i+1) 이후 코드를 실행함.(9-->1)

```
import time


# 반복적으로 구현한 n!
def factorial_iterative(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# 재귀적으로 구현한 n!
def factorial_recursive(n):
    if n <= 1:
        return 1
    # n! = n * (n - 1)!를 그대로 코드로 작성
    return n * factorial_recursive(n - 1)


start = time.time()
print('Iterative:', factorial_iterative(500))
print('time:', time.time() - start)
print()

start = time.time()
print('Recursive:', factorial_recursive(500))
print('time:', time.time() - start)

# n이 커질수록 재귀적으로 구현한 팩토리얼 계산이 더 느림.
```

```
# 유클리드 호제법
# A와 B의 최대공약수 = B와 R의 최대공약수
# R은 A를 B로 나눈 나머지 (A > B)

def gcd(a, b):
    if a%b == 0:
        return b
    return gcd(b, a%b)

gcd(120, 84)

#   gcd(120, 84) --> return gcd(84, 36) --> return gcd(36, 12) --> return 12
#                                                                       |
#   gcd(120, 84) = 12 <-- gcd(84, 36) = 12 <-- gcd(36, 12) = 12  <------
# 
```

# DFS

image.png

DFS 탐색 순서가 스택의 입출력 순서와 같음.\
**방문하지 않은 인접 노드가 없으면** 스택에서 최상단 노드를 **꺼내는 것**에 유의.

image.png

시작 노드인 '1'을 <스택>에 삽입하고 방문 처리\
탐색 순서 : 1 - 2 - 7 - 6 - 8 - 3 - 4 - 5

```
# DFS 동작 예시 구현

def dfs(graph, v, visited):
    # 현재 노드를 방문 처리
    visited[v] = True   # v는 시작 노드
    print(v, end = ' ')
    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        if not visited[i]: # i가 방문되지 않았다면 = i가 False
            dfs(graph, i, visited)

# 각 노드가 연결된 정보 표현 (2차원 리스트)
graph = [
         [],
         [2, 3, 8], # 1번 노드에 연결된 노드는 2번, 3번, 8번 노드
         [1, 7],    # 2번                   ''
         [1, 4, 5], 
         [3, 5],    # 숫자가 작은 노드 먼저 방문하는 조건은
         [3, 4],    # 왼쪽 graph에서 작은 숫자를 왼쪽으로 놓음으로써 구현
         [7],
         [2, 6, 8],
         [1, 7]
]

# 각 노드가 방문된 정보를 표현 (1차원 리스트)
visited = [False] * 9   # 0번째 인덱스는 사용하지 않음.

# 정의된 DFS 함수 호출
dfs(graph, 1, visited)
```

# BFS

image.png

BFS 탐색 순서가 큐의 입출력 순서와 같음.\
**큐에서 노드를 꺼낸 뒤에** 인접 노드를 큐에 삽입하는 것에 유의.

image.png

시작 노드인 '1'을 <큐>에 삽입하고 방문 처리\
탐색 순서: 1 2 3 8 7 4 5 6


```
# BFS 동작 예시 구현
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    # 현재 노드를 방문 처리
    visited[start] = True
    # 큐가 빌 때까지 반복
    while queue:
        # 큐에서 하나의 원소를 뽑아 출력
        v = queue.popleft()
        print(v, end = ' ')
        # 아직 방문하지 않은 인접한 원소들을 큐에 삽입
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

# 각 노드가 연결된 정보를 표현 (2차원 리스트)
graph = [
         [],
         [2, 3, 8], # 1번 노드에 연결된 노드는 2번, 3번, 8번 노드
         [1, 7],    # 2번                   ''
         [1, 4, 5], 
         [3, 5],    # 숫자가 작은 노드 먼저 방문하는 조건은
         [3, 4],    # 왼쪽 graph에서 작은 숫자를 왼쪽으로 놓음으로써 구현
         [7],
         [2, 6, 8],
         [1, 7]
]

# 각 노드가 방문된 정보를 표현 (1차원 리스트)
visited = [False] * 9   # 0번째 인덱스는 사용하지 않음.

# 정의된 BFS 함수 호출
bfs(graph, 1, visited)
```

```
# <문제> 음료수 얼려 먹기: 답안 예시 (DFS 이용)
def dfs(x, y):
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x-1, y)
        dfs(x, y-1)
        dfs(x+1, y)
        dfs(x, y+1)
        return True
    return False # elif graph[x][y]: return False 와 같음

n, m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int, input())))

# 모든 노드에 대해 음료수 채우기
result = 0
for i in range(n):
    for j in range(m):
        if dfs(i, j) == True:
            result += 1     # dfs(i, j) == True면, (i, j)와 연결된 
                            # 모든 노드에 대해서도 한꺼번에 True
print(result)
for i in range(n):
    for j in range(m):
        print(graph[i][j], end=' ')
    print()
```
